### **Introduction to SciPy**

### **1. What is SciPy?**
- **Definition:** SciPy is an open-source scientific computing library in Python, built on top of **NumPy**.
- **Purpose:** It provides a wide range of functions for mathematical calculations, scientific computing, and engineering problems.
- **Use Cases:** Commonly used in fields such as physics, chemistry, biology, engineering, and finance.

### **2. Overview of SciPy**
- **Core Features:**
  - Advanced mathematical functions (e.g., optimization, integration).
  - Special functions (e.g., Bessel functions).
  - Statistical distributions and tests.
  - Linear algebra functions (e.g., matrix operations).

### **5. Why Use SciPy?**
- **Efficiency:** Faster computations compared to manual implementations.
- **Rich Functionality:** Offers tools not available in NumPy.

### **7. Example: Solving a Differential Equation**

#### **Problem Statement:**
Model the motion of a falling object under gravity. The equation is:
\[ \frac{d^2y}{dt^2} = -g \]
Where \( g = 9.81 \, m/s^2 \).

#### Import Libraries:

````python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
````

### Define the Differential Equation:

````python
def model(y, t):
    g = 9.81  # acceleration due to gravity (m/s^2)
    return [y[1], -g]  # dy/dt = y[1], dy1/dt = -g
````

### Set Initial Conditions:

````python
y0 = [100, 0]  # Initial height (100 m) and initial velocity (0 m/s)
````

### Define Time Points:

````python
t = np.linspace(0, 5, 100)  # Time from 0 to 5 seconds
````

### Solve the Differential Equation:

````python
solution = odeint(model, y0, t)
````

### Ploting The Result

````python
plt.plot(t, solution[:, 0], label='Height (m)')
plt.title('Height of Falling Object Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.axhline(0, color='gray', lw=0.5, ls='--')  # Ground line
plt.legend()
plt.grid()
plt.show()
````

### **Understanding the Example:**
- The above code simulates the falling motion of an object under the influence of gravity.
- The plot visually represents how the height decreases over time until it reaches the ground.
